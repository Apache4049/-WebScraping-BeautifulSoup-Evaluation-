```python
import requests
from bs4 import BeautifulSoup

def scrape_odi_teams():
    url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    teams_data = []
    table = soup.find("table", class_="table")
    rows = table.find("tbody").find_all("tr")[:10]

    for row in rows:
        cols = row.find_all("td")
        team = cols[1].text.strip()
        matches = cols[2].text.strip()
        points = cols[3].text.strip()
        rating = cols[4].text.strip()
        teams_data.append({"team": team, "matches": matches, "points": points, "rating": rating})

    return teams_data

def scrape_odi_batsmen():
    url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    batsmen_data = []
    table = soup.find("table", class_="table")
    rows = table.find("tbody").find_all("tr")[:10]

    for row in rows:
        cols = row.find_all("td")
        player = cols[1].text.strip()
        team = cols[2].text.strip()
        rating = cols[3].text.strip()
        batsmen_data.append({"player": player, "team": team, "rating": rating})

    return batsmen_data

def scrape_odi_bowlers():
    url = "https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    bowlers_data = []
    table = soup.find("table", class_="table")
    rows = table.find("tbody").find_all("tr")[:10]

    for row in rows:
        cols = row.find_all("td")
        player = cols[1].text.strip()
        team = cols[2].text.strip()
        rating = cols[3].text.strip()
        bowlers_data.append({"player": player, "team": team, "rating": rating})

    return bowlers_data

if __name__ == "__main__":
    print("Top 10 ODI Teams in Men's Cricket:")
    teams = scrape_odi_teams()
    for idx, team in enumerate(teams, start=1):
        print(f"{idx}. {team['team']} - Matches: {team['matches']}, Points: {team['points']}, Rating: {team['rating']}")

    print("\nTop 10 ODI Batsmen:")
    batsmen = scrape_odi_batsmen()
    for idx, batsman in enumerate(batsmen, start=1):
        print(f"{idx}. {batsman['player']} ({batsman['team']}) - Rating: {batsman['rating']}")

    print("\nTop 10 ODI Bowlers:")
    bowlers = scrape_odi_bowlers()
    for idx, bowler in enumerate(bowlers, start=1):
        print(f"{idx}. {bowler['player']} ({bowler['team']}) - Rating: {bowler['rating']}")
```
